```python
 import pandas as pd
from google.colab import files

uploaded = files.upload()
data = pd.read_csv('titanic.csv')
data.head()

print(data.isnull().sum())

# Fill missing values with mean, median, and mode
data['Age'] = data['Age'].fillna(data['Age'].mean())  # Fill with mean
data['Age'] = data['Age'].fillna(data['Age'].median()) # Fill remaining with median
data['Age'] = data['Age'].fillna(data['Age'].mode()[0]) # Fill remaining (if any) with mode

print(data.isnull().sum())  # Verify the changes
```
